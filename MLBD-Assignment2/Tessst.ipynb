```python
import pandas as pd
df = pd.read_csv('glass.csv', header=None, names=["one","two","three","four","five","six","seven","eight","target"])
df.head()
```

```python
features = []
for feature in df.columns:
    if feature != 'target':
        features.append(feature)
X = df[features]
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
import matplotlib.pyplot as plt
count = y_train.value_counts()
count.plot.bar()
plt.ylabel('Number of records')
plt.xlabel('Target Class')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from scikitplot.metrics import plot_roc
from scikitplot.metrics import plot_precision_recall

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

y_score = model.predict_proba(X_test)
y_pred = model.predict(X_test)

#print(y_test)
#print(type(y_test))
print("y_test.shape", y_test.shape, "len", len(y_test), "type", type(y_test))
#print(y_score)
print("y_score.shape", y_score.shape, "len", len(y_score), "type", type(y_score))

# Plot metrics 
plot_roc(y_test, y_score)
plt.show()
    
plot_precision_recall(y_test, y_score)
plt.show()
```
